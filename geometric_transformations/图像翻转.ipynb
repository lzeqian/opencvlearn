```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
image=cv2.imread("../images/demo1.png")
rgbimg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgbimg)
plt.show()
```

```python
dst=cv2.flip(image,1)  #0表示延x轴翻转，正数沿着y轴翻转，负数同时沿着x和y轴翻转
rgbimg = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(rgbimg)
plt.show()
```

```python
dst=cv2.flip(image,0)  #0表示延x轴翻转，正数沿着y轴翻转，负数同时沿着x和y轴翻转
rgbimg = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(rgbimg)
plt.show()

```

```python
dst=cv2.flip(image,-1)  #0表示延x轴翻转，正数沿着y轴翻转，负数同时沿着x和y轴翻转
rgbimg = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(rgbimg)
plt.show()
```
