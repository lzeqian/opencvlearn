```python
import cv2
image=cv2.imread("../images/demo1.png")
height, width, channels = image.shape
step=4
for i in range(0, height, step):
    cv2.rectangle(image, (0, i), (width, i+step), (255, 255, 255), 1)

for j in range(0, width, step):
    cv2.rectangle(image, (j, 0), (j+step, height), (255, 255, 255), 1)

cv2.imshow("image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 这里每4个像素绘制一个正方形的效率要比opencv慢很多
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# 读取图像
img = plt.imread("../images/demo1.png")

# 绘制图像
plt.imshow(img)

# 绘制正方形边框
for i in range(0, img.shape[0], 4):
    for j in range(0, img.shape[1], 4):
        rect = Rectangle((j, i), 4, 4, linewidth=1, edgecolor='w', facecolor='none')
        plt.gca().add_patch(rect)

# 显示图像和边框
plt.show()
```

```python
#获取某个像素点的值
import cv2
image=cv2.imread("../images/demo1.png")
px=image[291,218] #注意坐标在opencv是反的 y,x
print(px)  #输出的通道是bgr
blue = image[291, 218, 0]         # 坐标(291, 218)上的像素的B通道的值
green = image[291, 218, 1]        # 坐标(291, 218)上的像素的G通道的值
red = image[291, 218, 2]          # 坐标(291, 218)上的像素的R通道的值
print(blue, green, red)
image[291,218]=[255,255,255]   #将这个点上的图像设置成255，255,255是白色的
print(image[291,218])  #输出的通道是bgr
```
