```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from myutils.common import show,fillColor
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
matplotlib.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
img=cv2.imread("./images/1.jpg")
grayImg=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(grayImg, "卷积",cmap="gray",debug=True)
# 颜色范围
lower = np.array(0)
upper = np.array(50)

# 创建卷积核
kernel = np.ones((9, 9), dtype=np.float32)

# 进行卷积操作
result = cv2.filter2D(grayImg, -1, kernel)
show(result, "卷积",cmap="gray",debug=True)
# 将在颜色范围内的像素设为255，不在颜色范围内的像素设为0
result = cv2.inRange(result, lower, upper)
show(result, "inRange",cmap="gray",debug=True)
```
