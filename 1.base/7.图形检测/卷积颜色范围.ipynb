```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from myutils.common import show,fillColor
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 
matplotlib.rcParams['axes.unicode_minus'] = False  # 
img=cv2.imread("./images/1.jpg")
grayImg=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(grayImg, "",cmap="gray",debug=True)
# 
lower = np.array(0)
upper = np.array(50)

# 
kernel = np.ones((9, 9), dtype=np.float32)

# 
result = cv2.filter2D(grayImg, -1, kernel)
show(result, "",cmap="gray",debug=True)
# 2550
result = cv2.inRange(result, lower, upper)
show(result, "inRange",cmap="gray",debug=True)
```
