```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from myutils import common
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 
matplotlib.rcParams['axes.unicode_minus'] = False  # 
img=cv2.imread("./idcard2.png")
# 5
kernel_size = 5
# blur_img = cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)
grayImg=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
common.show(grayImg,"",cmap="gray",debug=True)
```

```python
"""
CannyJohn F. Canny1986
OpenCVCannyCanny()
     edges = cv2.Canny(image, threshold1, threshold2, apertureSize, L2gradient)


image
threshold1
threshold2
apertureSizeSobel
L2gradientFalseTrue

edges
22
"""
# 
# 
low_threshold = 50
high_threshold = 150
sobelx = cv2.Sobel(grayImg, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(grayImg, cv2.CV_64F, 0, 1, ksize=3)
gradient_mag = cv2.magnitude(sobelx, sobely)
gradient_dir = cv2.phase(sobelx, sobely, angleInDegrees=True)
# 0255
gradient_mag = cv2.normalize(gradient_mag, None, 0, 255, cv2.NORM_MINMAX)

# CV_8U
gradient_mag = gradient_mag.astype('uint8')
edges = cv2.Canny(gradient_mag, low_threshold, high_threshold, apertureSize=3, L2gradient=True)


# edges = cv2.Canny(grayImg, 20, 40)
# edges = cv2.Canny(grayImg, 50, 200)
common.show(edges,"",cmap="gray",debug=True)
zeroArray=np.ones((12,12))
pzt=cv2.dilate(edges, zeroArray)
common.show(pzt,"",cmap="gray",debug=True)
# dest=cv2.blur(pzt,(9,9))
zeroArray=np.ones((2,2))
dst = cv2.morphologyEx(pzt, cv2.MORPH_GRADIENT, zeroArray)
common.show(dst,"",cmap="gray",debug=True)
contours, hierarchy = cv2.findContours(dst, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
common.show(cv2.drawContours(img.copy(), contours, -1, (0, 0, 255), 1),"",cmap="gray",debug=True)
contours = sorted(contours, key=cv2.contourArea, reverse=True)[0]
common.show(cv2.drawContours(img.copy(), [contours], -1, (0, 0, 255), 1),"",cmap="gray",debug=True)
```
