```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from utils.common import show,fillColor
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
matplotlib.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
img=cv2.imread("./images/idcard1.jpg")
grayImg=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
show(grayImg,"原始图",debug=True)
lower_orange_dish = np.array([0,   0, 0])#颜色范围低阈值
upper_orange_dish= np.array([200,  200, 200])#颜色范围高阈值
#cv2.inRange函数来根据范围进行二值化。该函数需要输入原始图像和两个颜色值作为参数，
#然后将图像中在这两个颜色值范围内的像素设置为白色（255），其余的像素设置为黑色（0）。
mask_orange_dish = cv2.inRange(img,lower_orange_dish,upper_orange_dish)#根据颜色范围删选
mask_orange_dish = cv2.medianBlur(mask_orange_dish, 5)#中值滤波
show(mask_orange_dish,"原始图",cmap="gray",debug=True)
```
