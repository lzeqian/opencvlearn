```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from myutils import common


matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 
matplotlib.rcParams['axes.unicode_minus'] = False  # 
img=cv2.imread("./images/1.jpg")
grayImg=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
common.show(grayImg,"",cmap="gray",debug=True)
edges = cv2.Canny(grayImg, 20, 40)
common.show(edges,"",cmap="gray",debug=True)
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 15, 100, 18)
img1=img.copy()
for line in lines:
    x1,y1,x2,y2=line[0]
    cv2.line(img1,(x1,y1),(x2,y2),(0,0,255),2)
common.show(img1,"",cmap="gray",debug=True)    
```
