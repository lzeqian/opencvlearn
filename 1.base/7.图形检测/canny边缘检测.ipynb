```python

import cv2
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 
matplotlib.rcParams['axes.unicode_minus'] = False  # 
img=cv2.imread("./images/idcard1.jpg")
plt.subplot(221)
plt.title("")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#
r1=cv2.Canny(img,10,50)
r2=cv2.Canny(img,100,200)
r3=cv2.Canny(img,400,600)
plt.subplot(222)
plt.imshow(cv2.cvtColor(r1, cv2.COLOR_BGR2RGB))
plt.subplot(223)
plt.imshow(cv2.cvtColor(r2, cv2.COLOR_BGR2RGB))
plt.subplot(224)
plt.imshow(cv2.cvtColor(r3, cv2.COLOR_BGR2RGB))
```
