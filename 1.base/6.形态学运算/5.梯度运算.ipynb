```python
"""


12.15
"""
import matplotlib.pyplot as plt
import matplotlib
import cv2
import numpy as np
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 
matplotlib.rcParams['axes.unicode_minus'] = False  # 
image=cv2.imread("./images/2.jpg");
zeroArray=np.ones((5,5))
plt.subplot(221)
plt.title("")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(224)
plt.subplot(222)
plt.title("")
dst = cv2.morphologyEx(image, cv2.MORPH_ERODE, zeroArray) #erode
plt.imshow(dst)
plt.subplot(223)
plt.title("")
dst = cv2.morphologyEx(image, cv2.MORPH_DILATE, zeroArray) #erode
plt.imshow(dst)
plt.subplot(224)
plt.title("")
dst = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, zeroArray) #erode
plt.imshow(dst)
```
