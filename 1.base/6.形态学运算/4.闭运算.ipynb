```python
"""

"""
import matplotlib.pyplot as plt
import matplotlib
import cv2
import numpy as np
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 
matplotlib.rcParams['axes.unicode_minus'] = False  # 
image=cv2.imread("./images/2.jpg");
plt.subplot(221)
plt.title("")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(222)
plt.title("")
zeroArray=np.ones((5,5))
#
dest=cv2.dilate(image, zeroArray)
dest=cv2.erode(dest, zeroArray)
plt.imshow(dest)
plt.subplot(224)
plt.title("")
dst = cv2.morphologyEx(image, cv2.MORPH_CLOSE, zeroArray) #erode
plt.imshow(dst)
```
