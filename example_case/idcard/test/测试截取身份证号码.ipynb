```python
%matplotlib inline
import os
import cv2
from example_case.idcard.card import Card
from utils import common
from utils.common import get_all_files


dir_path = os.getcwd() + "/../images"
all_files = get_all_files(dir_path)
print(all_files)
for i in all_files:
    card = Card(i, isDebug=False)
    cobj=card.getCardNoLocation()
    if cobj:
        x, y, w, h=cobj
        imgc=card.img.copy()
        cv2.rectangle(imgc, (x, y), (x + w, y + h), (0, 0, 255), 1)
        common.show(imgc, "边框图", cmap="gray", debug=True)
```
