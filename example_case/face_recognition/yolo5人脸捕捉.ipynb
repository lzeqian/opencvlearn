```python
import cv2
import matplotlib.pyplot as plot
%matplotlib inline
import cv2
import torch
import matplotlib.pyplot as plt

# model = torch.load('ultralytics/yolov5', 'custom', path='./best.pt', force_reload=True)
model = torch.hub.load('D:\\code1\\yolov5-master\\yolov5-master', 'custom', path='../idcard/yolov5/bestface.pt', source='local')  # local repo
#print(model)
# 读取图像
img = cv2.imread('./images/rxt.jpg')
# 进行预测
results = model(img)
resultLabel=[]
print(model.names)
# 解析预测结果
for result in results.xyxy[0]:
    x1, y1, x2, y2, conf, cls = result.tolist()
    if conf > 0.1:
        # 绘制边框和标签
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv2.putText(img, f"{model.names[int(cls)]} {conf:.2f}", (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        resultLabel.append(model.names[int(cls)])
# 显示图像
print("预测的结果是",resultLabel)
plt.imshow(img)
plt.show()
```
