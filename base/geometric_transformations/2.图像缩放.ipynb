```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
matplotlib.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
image=cv2.imread("../../images/demo1.png")
dst=cv2.resize(image,(100,100))  #直接根据大小缩放
plt.subplot(121)
plt.xlim(0, 800)  #x轴限制显示1-800
plt.ylim(0, 400)  #y轴限制显示1-800  固定轴的大小才能看出比例的变化
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("原图")
plt.subplot(122)
plt.xlim(0, 800);plt.ylim(0, 400)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.title("根据大小缩放")
plt.show()
```

```python
import cv2
import numpy as np
image=cv2.imread("../../images/demo1.png")
dst=cv2.resize(image,None,fx=0.5,fy=0.5)  #直接根据大小缩放,缩放50%
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("原图")
plt.subplot(122)
plt.xlim(0, 800);plt.ylim(0, 400)
plt.title("根据比例缩放")
plt.xlim(0, 800);plt.ylim(0, 400)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
