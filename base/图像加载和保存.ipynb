````python
import cv2
import matplotlib.pyplot as plt
#加载图像,使用matplotlib输出
bgrimg=cv2.imread("../images/demo1.png")
# 将BGR格式转换为RGB格式
rgbimg = cv2.cvtColor(bgrimg, cv2.COLOR_BGR2RGB)
plt.imshow(rgbimg)
plt.show()
````

````python
#加载图像，使用cv2编辑器显示
bgrimg=cv2.imread("../images/demo1.png")
cv2.imshow('demo',bgrimg)
# delay：等待用户按下键盘上按键的时间，单位为毫秒（ms）。当delay的值为负数、0或者空时，表示无限等待用户按下键盘上按键的时间。
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
#读取图片,写入图像。
img=cv2.imread("../images/demo.png")
cv2.imwrite("../images/demo_write.png",img)
````
